```python
import requests
import bs4
from time import sleep
from random import randint
import pandas as pd
```

```python
# URL
url = 'https://www.imdb.com/title/tt5956100/?ref_=fn_al_tt_1'
```

```python
r = requests.get(url)
```

```python
soup = bs4.BeautifulSoup(r.content)
```

```python
details = soup.find('div' , attrs = {'id' : 'titleDetails'})
```

```python
details_divs = details.find_all('div' , attrs = {'class' : 'txt-block'})
movie_details = []
i1 = 0
for det in details_divs:
    if(i1==11):
        break
    heading = det.find('h4').text
    movie_details.append([heading ,det.text.replace(heading,'')])
    i1 = i1 + 1
```

```python
for i in movie_details:
    print (i[0] +' -----------> '+i[1].replace('\n',''))
```

```python
from selenium import webdriver 
```

```python
driver = webdriver.Chrome('./chromedriver')
```

```python
url = 'https://www.imdb.com/title/tt5956100/reviews?spoiler=hide&sort=helpfulnessScore&dir=desc&ratingFilter=0'
driver.get(url)
```

```python
def Load_All():
    i = 0
    num = 100
    for div in driver.find_elements_by_class_name('lister-item'):
        i = i+1
    if(i<num):
        print('List -> ' + str(i))
        driver.find_element_by_class_name('ipl-load-more__button').click()
        sleep(randint(2,3))
        Load_All()
    else:
        print('Done ------------> ' + str(i))
        return

def Extract_Reviews():
    i = 0   
    rating = []
    titl = []
    dat = []
    conten = []
    for div in driver.find_elements_by_class_name('lister-item'):       
        i = i+1
        try:
            div = div.find_element_by_class_name('review-container')
            # Ratings
            ratings = div.find_element_by_class_name('ipl-ratings-bar').text
            # Title
            title = div.find_element_by_class_name('title').text
            # Date
            date = div.find_element_by_class_name('review-date').text
            # Content
            content = div.find_element_by_class_name('content').find_element_by_class_name('text').text
            
            print(ratings)
            print(title)
            print(date)
            
            rating.append(ratings)
            titl.append(title)
            dat.append(date)
            conten.append(content)
            print('Done ----> ' + str(i))
        except:
            print("Error Occured : " + str(i))
    return {'Ratings' : rating , 'Title' : titl , 'Date' : dat , 'Content' : conten}            
            
def MakeDataframe(d):
    df = pd.DataFrame()
    df['Ratings'] = d['Ratings']
    df['Title'] = d['Title']
    df['Date'] = d['Date']
    df['Content'] = d['Content']
    return df
```

```python
Load_All()
reviews = Extract_Reviews()
```

```python
df = MakeDataframe(reviews)
df.to_csv('TigerZindaHai1.txt',sep=',')
print('Done ---------------------- :)')
```
